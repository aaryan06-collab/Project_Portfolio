```python
def prediction():
    try:
        from sklearn.preprocessing import LabelEncoder
        df=pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\All jupyter projects\\loan_data.csv")
        x=df.drop("Loan_Status",axis='columns')
        y=df['Loan_Status']
        le=LabelEncoder()
        x['gender']=le.fit_transform(x['Gender'])
        x['married']=le.fit_transform(x['Married'])
        x['education']=le.fit_transform(x['Education'])
        x['self_employed']=le.fit_transform(x['Self_Employed'])
        x_c=x.drop(['Gender','Married','Education','Self_Employed'],axis='columns')
        model=dt(class_weight='balanced')
        model.fit(x_c,y)
        inc=int(ei.get())
        lna=int(el.get())
        crh=int(bool(sc.get()))
        gnd=sg.get()
        if gnd=='Male':
            gndn=1
        else:
            gndn=0
        mr=int(bool(sm.get()))
        grd=int(bool(se.get()))
        sle=int(bool(ss.get()))
        g=model.predict([[inc,lna,crh,gndn,mr,grd,sle]])
        if g[0]=="Approved":
            pre.config(text='Congrats. Your loan will be approved ✅',font=('bold',15),bg='white')
        else:
            pre.config(text='Sorry. Your loan will not be approved ❌',font=('bold',15),bg='white')
    except:
        mb.showerror("Error", "Please fill all the fields properly.")
```
