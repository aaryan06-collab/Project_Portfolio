```python
from tkinter import *
import random
from tkinter import filedialog as f
import os
import pygame
from tkinter import messagebox         

ms=Tk()
ms.title('Music Player')
ms.geometry('600x500')
#ms.iconbitmap(r"C:\Users\Lenovo\Desktop\All jupyter projects\m.ico")
ms.resizable(0,0)
pygame.mixer.init()

songlist=Listbox(ms,width=100,height=20,bg='black',fg='white')
songlist.pack()

songs=[]
csong=''
dire='none'
paused=False
def select():
    global csong,dire
    try:
        dire=f.askdirectory()
        for i in os.listdir(dire):
            n,e=os.path.splitext(i)
            if e=='.mp3':
                songs.append(i)
        for i in songs:
            songlist.insert(END,i)
        songlist.selection_set(0)
        csong=songs[songlist.curselection()[0]]
    except:
        messagebox.showerror("Error", "No directory choosen.")

def on_close():
    conf=messagebox.askquestion("Confirmation", "Do you want to exit?")
    if conf=='yes':
        pygame.mixer.music.stop()
        ms.destroy()

def select_song(event):
    try:
        global csong
        csong = songlist.get(ANCHOR)
        play()
    except:
        pass
songlist.bind('<<ListboxSelect>>', select_song)
    
def play():
    try:
        global csong,paused,dire
        if not paused:
            pygame.mixer.music.load(os.path.join(dire,csong))
            pygame.mixer.music.play()
        else:
            pygame.mixer.music.pause()
            paused=False
    except:
        messagebox.showerror("Error", "Invalid song.")
       
def pause():
    try:
        global paused
        pygame.mixer.music.pause()
        pause=True
    except:
        messagebox.showinfo("Information","Choose a valid playlist.")
        
def shuffle_song():
    try:
        global csong
        c_index=songs.index(csong)
        new_index=c_index
        while new_index==c_index:
            new_index = random.randint(0, len(songs) - 1)
        csong=songs[new_index]
        play()
    except:
        pass

def nextm():
    global csong,songlist
    try:
        songlist.selection_clear(0,END)
        songlist.selection_set(songs.index(csong)+1)
        csong=songs[songlist.curselection()[0]]
        play()
    except:
        if csong==songs[-1]:
            csong=songs[0]
            play()

def previous():
    global csong,songlist
    try:
        songlist.selection_clear(0,END)
        songlist.selection_set(songs.index(csong)-1)
        csong=songs[songlist.curselection()[0]]
        play()
    except:
        if csong==songs[0]:
            csong=songs[-1]
            play()
  
mbar=Menu(ms)
ms.config(menu=mbar)
omenu=Menu(mbar, tearoff='False')
omenu.add_command(label='Select Folder',command=select)
mbar.add_cascade(label='Organise', menu=omenu)

ms.protocol('WM_DELETE_WINDOW',on_close)
try:
    play_image=PhotoImage(file=r"C:\\Users\\Lenovo\\Desktop\\All jupyter projects\\play.png")
    pbn=Button(ms, image=play_image, borderwidth=0, command=play)
    pbn.place(x=220,y=400)
    pause_image=PhotoImage(file=r"C:\\Users\Lenovo\\Desktop\\All jupyter projects\\pause.png")
    pabn=Button(ms, image=pause_image, borderwidth=0, command=pause)
    pabn.place(x=300,y=400)
    next_image=PhotoImage(file=r"C:\\Users\\Lenovo\Desktop\\All jupyter projects\\next.png")
    nbn=Button(ms, image=next_image, borderwidth=0, command=nextm)
    nbn.place(x=380,y=400) 
    prev_image=PhotoImage(file=r"C:\\Users\Lenovo\\Desktop\\All jupyter projects\\previous.png")
    rbn=Button(ms, image=prev_image, borderwidth=0, command=previous)
    rbn.place(x=160,y=400)
    shuffle_image = PhotoImage(file=r"C:\\Users\\Lenovo\Desktop\\All jupyter projects\\shuffle.png")
    shbn=Button(ms,image=shuffle_image,borderwidth=0,command=shuffle_song)
    shbn.place(x=10,y=400)
except:
    pass

ms.mainloop()
```
