```python
from tkinter import *
import pandas as pd 
import numpy as np
from sklearn import linear_model as lm
ar=bm=ge=None
global ar,bm,ge
```

```python
def pr():
    try:
        a=ar.get()
        b=bm.get()
        g=ge.get()
        if a == "" or b == "" or g == "":
            r.config(text="Please fill all fields!")
        a = int(a)
        b = int(b)
        g = int(g)
        df=pd.read_excel("C:\\Users\\Lenovo\\Desktop\\All jupyter projects\\prices.xlsx")
        x=df.drop('price',axis='columns')
        y=df['price']
        reg=lm.LinearRegression()
        reg.fit(x,y)
        p=reg.predict([[a,b,g]])
        r.config(text="₹ " + str(float(p)))
    except:
        pass
```

```python
ms=Tk()
ms.title('Plot price predictor')
ms.geometry('300x420')
ms.resizable(0,0)


w=Label(ms,text="Warning: Only digits are allowed",font=(12))
w.pack()

ab=Label(ms,text='Area')
ab.place(x=30,y=60)
ar=Entry(ms)
ar.place(x=100,y=60)

bd=Label(ms,text='Bedroom')
bd.place(x=30,y=120)
bm=Entry(ms)
bm.place(x=100,y=120)

ag=Label(ms,text='Age')
ag.place(x=30,y=180)
ge=Entry(ms)
ge.place(x=100,y=180)

btn=Button(ms,text='Predict Price',command=pr)
btn.place(x=90,y=250)

pp=Label(ms,text='Predicted Price:',font=(12))
pp.place(x=0,y=310)
r=Label(ms,text='',font=(14))
r.place(x=70,y=350)



ms.mainloop()
```
