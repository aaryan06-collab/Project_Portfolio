```python
from tkinter import *
from tkinter import messagebox

firstn=ope=secn=None

def g_d(digit):
    c=n1['text']
    new=c+str(digit)
    n1.config(text=new)

def clear():
    n1.config(text='') 
    
def oper(op):
    global firstn,ope
    firstn=eval(n1['text'])
    ope=op
    n1.config(text='')
    
def res():
    global firstn,ope,secn
    secn=eval(n1['text'])
    if ope=='+':
        n1.config(text=str(firstn+secn))
    elif ope=='-':
        n1.config(text=str(firstn-secn))
    elif ope=='*':
        n1.config(text=str(firstn*secn))
    else:
        if secn==0:
            messagebox.showerror("Error", "Cannot divided by zero.")
        else:
            n1.config(text=str(firstn/secn))

def bs():
    dele= n1['text']
    n1.config(text=dele[:-1])   

top = Tk()
top.title('Calculator')
top.geometry('240x400')
top.resizable(0,0)

n1=Label(top,text='')
n1.place(x=0,y=80)
n1.config(font=(80))

nb7=Button(top,text='7',width=6,height=3,command=lambda :g_d(7))
nb7.place(x=0,y=160,)
nb8=Button(top,text='8',width=6,height=3,command=lambda :g_d(8))
nb8.place(x=60,y=160)
nb9=Button(top,text='9',width=6,height=3,command=lambda :g_d(9))
nb9.place(x=120,y=160)
nba=Button(top,text='+',width=6,height=3,command=lambda :oper('+'))
nba.place(x=180,y=160)

nb4=Button(top,text='4',width=6,height=3,command=lambda :g_d(4))
nb4.place(x=0,y=220)
nb5=Button(top,text='5',width=6,height=3,command=lambda :g_d(5))
nb5.place(x=60,y=220)
nb6=Button(top,text='6',width=6,height=3,command=lambda :g_d(6))
nb6.place(x=120,y=220)
nbs=Button(top,text='-',width=6,height=3,command=lambda :oper('-'))
nbs.place(x=180,y=220)

nb1=Button(top,text='1',width=6,height=3,command=lambda :g_d(1))
nb1.place(x=0,y=280)
nb2=Button(top,text='2',width=6,height=3,command=lambda :g_d(2))
nb2.place(x=60,y=280)
nb3=Button(top,text='3',width=6,height=3,command=lambda :g_d(3))
nb3.place(x=120,y=280)
nbm=Button(top,text='*',width=6,height=3,command=lambda :oper('*'))
nbm.place(x=180,y=280)

nb0=Button(top,text='0',width=15,height=3,command=lambda :g_d(0))
nb0.place(x=0,y=340)
nbd=Button(top,text='.',width=6,height=3,command=lambda :g_d('.'))
nbd.place(x=120,y=340)
nbe=Button(top,text='=',width=6,height=3,command=lambda :res())
nbe.place(x=180,y=340)

nbc=Button(top,text='Clear',width=15,height=3,command=lambda :clear())
nbc.place(x=0,y=100)
nbbs=Button(top,text='back',width=6,height=3,command=bs)
nbbs.place(x=120,y=100)
nbdi=Button(top,text='/',width=6,height=3,command=lambda :oper('/'))
nbdi.place(x=180,y=100)


top.mainloop()
```
